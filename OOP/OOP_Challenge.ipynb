# Object Oriented Programming Challenge

For this challenge, create a bank account class that has two attributes:

* owner
* balance

and two methods:

* deposit
* withdraw

As an added requirement, withdrawals may not exceed the available balance.

Instantiate your class, make several deposits and withdrawals, and test to make sure the account can't be overdrawn.

```python
class Account:
    
    def __init__(self, owner, balance=0.0):
        self.owner = owner
        self.balance = balance
    
    def __repr__(self):
        return f'{"Account Owner:":17s}{self.owner}\n{"Current balance:":17s}${self.balance}'
    
    def deposit(self, money):
        if self.__check_valid(money):
            self.balance += money
            print(f'Deposit Accepted: ${money}')
            return True
        else:
            print(f'Deposit Denied')
            return False
    
    def withdraw(self, money):
        if self.__check_valid(money):
            if self.balance < money:
                print(f'Withdraw Denied: Insufficent Balance')
                return False
            else:
                self.balance -= money
                print(f'Withdraw Accepted: ${money}')
                return True
        else:
            print(f'Withdraw Denied')
            return False
        
    # helper function
    def __check_valid(self, money):
        if isinstance(money, (int, float)):
            if money <= 0:
                print(f'Error: please type postivie number')
                return False
            else:
                return True
        else:
            print(f'Error: please type numeric number')
            return False
```

```python
# 1. Instantiate the class
my_account = Account('Jose', 100)

# 2. Print the object
print(my_account)
print('-' * 20)

# 3. Make a series of deposits 
my_account.deposit(50)
print(my_account)
print('-' * 20)
my_account.deposit(-10)
print(my_account)
print('-' * 20)

# 4. Make a series of withdrawals
my_account.withdraw(10)
print(my_account)
print('-' * 20)
my_account.withdraw(15)
print(my_account)
print('-' * 20)

# 5. Make a withdrawal that exceeds the available balance
my_account.withdraw(500)
print(my_account)
```
